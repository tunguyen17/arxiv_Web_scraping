```python
data_list = []
title_list = []
new_title = []
au_list = []
sub_art = []
for t in all_cat:#for all sub fields
    url_3 = 'http://export.arxiv.org/api/query?search_query=cat:'+ t + '&start=0&max_results=1'
    #shows first 1 papers for all subjects and sunfields
    rd_3 = urllib.request.urlopen(url_3)#open the url
    soup_3 = BeautifulSoup(rd_3, 'xml')#allows me to read the xml for the url given
    
    
    entry = soup_3.findAll('entry')#gind all the headers named entry and then output all the data inside
    #the entry headers
    for i in entry:#iterates through each entry header
        au_new = []

        for author in i.findAll("name"):#in each entry find all the names of the authors
            #and add them to the list called au_new

            au_new.append(author.get_text())
        au_list.append(au_new)#puts the whole au_new list into au_list before clearing au_new
        au_list
        for title in i.findAll('title'):#find all the titles in between each entry header 
            new_title.append(title.get_text())#only get the text between the headers
        sub_art.append(i.primary_category['term'])#picks the subject field of each article
    # display(cat_au)
    # display(title)
    # display(subject)
#     data = pd.concat([title,cat_au,subject],axis = 1)
#     display(data)
authors = pd.DataFrame(au_list)
titles = pd.DataFrame(new_title)
subjects = pd.DataFrame(sub_art)
use_data = pd.concat([titles,authors,subjects],axis = 1)
use_data
```

```python
url_list = 'http://export.arxiv.org/help/api/archive_list'
data_list = urllib.request.urlopen(url_list)
soup_list = BeautifulSoup(data_list, 'xml')
ul = soup_list.findAll('ul')
new_sub_list = []
for i in ul:
    b = i.findAll('b')
   
    for j in b:
        a = j.findAll('a')
        for x in a:
            new_sub_list.append(x.contents[0])
        
sub_data = pd.DataFrame(new_sub_list) 
sub_data
sub_data.rename(columns = {0:'subject'},inplace = True)
abv = ['cs','econ','eess', 'math', 'astro-ph','cond-ma', 'gr-qc', 'hep-ex','hep-la','hep-ph','hep-th','math-ph', 'nlin','nucl-ex', 'nucl-th', 'physics','quant-ph','q-bio','q-fin','stat']
abv = pd.DataFrame(abv)
abv.rename(columns = {0:'abv'},inplace = True)
sub_df = pd.concat([sub_data,abv],axis =1)

sub_df
```
